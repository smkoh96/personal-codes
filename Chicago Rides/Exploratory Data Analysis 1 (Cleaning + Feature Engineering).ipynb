### Data Cleaning

```python
main_df.isnull().mean()*100
```

### Removed rows with null values

```python
main_df=main_df.dropna(subset=['Trip Seconds','Trip Miles','Fare','Tips','Tolls','Extras','Trip Total'])
```

```python
main_df = main_df.fillna(value={'Pickup Community Area':'outside','Dropoff Community Area':'outside'})
main_df
```

### Convert to int type for Numeric Features

```python
main_df['Trip Seconds'] = [(str(i).replace(",", "")) for i in main_df['Trip Seconds']]
main_df['Trip Miles'] = [(str(i).replace(",", "")) for i in main_df['Trip Miles']]
main_df['Fare'] = [(str(i).replace(",", "")) for i in main_df['Fare']]

main_df = main_df.astype({"Trip Seconds": int, "Trip Miles": float,"Fare": float})
```

### Removed rows with zero value for Trip miles & Seconds

```python
main_df = main_df[main_df['Trip Miles']!=0]
main_df = main_df[main_df['Trip Seconds']!=0]
```

# Details on Free trips

```python
free_trips=main_df[main_df['Fare']==0]
```

```python
plt.figure(figsize=(20,6))
plt.title('Free Trip')
free_trips['Payment Type'].value_counts().plot(kind='bar',rot=0)
plt.show()

print(free_trips['Payment Type'].value_counts())
```

```python
plt.figure(figsize=(12,10))
plt.title('Companies that provide free trips')

free_trips_company=free_trips['Company'].value_counts()
sns.barplot(y= free_trips_company.index, x = free_trips_company.values, orient='h')
plt.show()
print(free_trips['Company'].value_counts())
```

```python
#remove free trips
main_df = main_df[main_df['Fare']!=0]
main_df.sample(10)[['Fare','Tips','Tolls','Extras','Trip Total']]
```

```python
main_df['Local_pickup']= np.where(main_df['Pickup Community Area'] !='outside', True, False)
```

```python
main_df['Local_dropoff']= np.where(main_df['Dropoff Community Area'] !='outside', True, False)
```

# too many abnormal data points, i.e. high miles but low duration

```python
plt.figure(figsize=(20,10))
plt.title('Scatter plot of distance vs time')
sns.scatterplot(x='Trip Miles',y='Trip Seconds',data=main_df)
plt.show()
```

```python
main_df['Miles per second'] = main_df['Trip Miles']/main_df['Trip Seconds']
```

```python
f, axes = plt.subplots(2,1, figsize=(15, 15))


sns.violinplot(x=main_df['Miles per second'], ax = axes[0])
#plt.title('Number of Passengers Aboard')

sns.boxplot(x=main_df['Miles per second'],ax = axes[1])
#plt.title('Number of Passengers Aboard')
plt.rcParams.update({'font.size': 15})
plt.show()

print(main_df['Miles per second'].describe())
```

### Removed outlier based on IQR

```python
Q1=main_df['Miles per second'].describe()['25%']
Q3=main_df['Miles per second'].describe()['75%']
IQR = Q3-Q1
print('UB:',Q3 + 1.5 * IQR)
print('LB:',Q1 - 1.5 * IQR)
```

```python
main_df.shape
```

```python
Q1=main_df['Miles per second'].describe()['25%']
Q3=main_df['Miles per second'].describe()['75%']
IQR = Q3-Q1
clean_df = main_df[ main_df['Miles per second']<(Q3 + 1.5 * IQR)]
```

```python
clean_df.shape
```

### Removed irrelevant Columns

```python
clean_df = clean_df.drop(columns=['Trip ID','Taxi ID','Pickup Community Area','Dropoff Community Area','Pickup Census Tract','Dropoff Census Tract'])
```

```python
clean_df.to_csv('Taxi_trip_cleaned.csv')
```
